```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("/Users/pintoza/Desktop/dev/data-science/sales_optimization/data/processed/df_final.csv")
```

```python
df.head()
```

```python
# Assuming df is your dataframe

# Pivot table to create the user-item matrix
user_item_matrix = df.pivot_table(index='customer_id', columns='order_number', values='unit_price', aggfunc='count').fillna(0)

# Convert any interaction to 1 (indicating interaction happened), and 0 otherwise
user_item_matrix = user_item_matrix.applymap(lambda x: 1 if x > 0 else 0)

user_item_matrix
```

```python
# Compute item similarity matrix
item_similarity = cosine_similarity(user_item_matrix.T) # Transpose to get items as rows
item_similarity_df = pd.DataFrame(item_similarity, index=user_item_matrix.columns, columns=user_item_matrix.columns)

item_similarity_df
```

```python
non_zero_pairs = item_similarity_df[item_similarity_df != 0].stack().rename_axis(('Item_A', 'Item_B')).reset_index(name='Similarity')
# Filter out the self-similarity scores (where Item_A is the same as Item_B)
non_zero_pairs = non_zero_pairs[non_zero_pairs['Item_A'] != non_zero_pairs['Item_B']]
non_zero_pairs
```

```python
plt.figure(figsize=(10,5))
plt.hist(non_zero_pairs['Similarity'], bins=30, edgecolor='black')
plt.title('Distribution of Non-Zero Similarity Scores')
plt.xlabel('Similarity Score')
plt.ylabel('Number of Item Pairs')
plt.grid(axis='y')
plt.show()
```

```python
common_items = non_zero_pairs['Item_A'].value_counts().head(10)
common_items
```
