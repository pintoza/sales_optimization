```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("/Users/pintoza/Desktop/dev/data-science/sales_optimization/data/processed/df_final.csv")
```

```python
df.head()
```

```python
customer_df = df.groupby('customer_id').agg({
    'unit_price': 'sum',
    'order_number': 'count',
    'order_date': 'max'
}).reset_index()
```

```python
df = customer_df
```

```python
df.head()
```

```python
df.rename(columns={
    'unit_price': 'Customer_Lifetime_Value',
    'order_number': 'Total_Orders',
    'order_date': 'Most_Recent_Order'
}, inplace=True)
```

```python
df.head()
```

```python
df['Average_Purchase_Frequency'] = df['Total_Orders'] / df['customer_id'].nunique()
```

```python
df.head()
```

```python
# Select numerical columns only
numerical_cols = ['Customer_Lifetime_Value', 'Total_Orders', 'Average_Purchase_Frequency']

# Create a scaler object
scaler = StandardScaler()

# Fit and transform the numerical features
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
df.head()
```

```python
# Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df[numerical_cols])
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Create KMeans object
kmeans = KMeans(n_clusters=4)

# Fit the model
df['Cluster'] = kmeans.fit_predict(df[numerical_cols])
```

```python
centroids = kmeans.cluster_centers_
centroids
```

```python
# Count of records in each cluster
cluster_distribution = df['Cluster'].value_counts().reset_index()
cluster_distribution
```

Cluster 1: High Lifetime Value, Frequent Purchase but High Recency

Cluster 2: Low Lifetime Value, Infrequent Purchase and High Recency

Cluster 3: High Lifetime Value, Frequent Purchase and Low Recency

Cluster 4: Low Lifetime Value, Infrequent Purchase but Low Recency

```python
# Filter out non-numeric columns
df_numeric = df.select_dtypes(include=[np.number])

# Apply PCA and transform the numeric data to 2D
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df_numeric)

# Add the cluster assignments to our reduced numeric data
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
principal_df['Cluster'] = kmeans.labels_

# Visualize the clusters
plt.figure(figsize=(10, 8))
colors = ['r', 'g', 'b', 'y']  # Add more colors if you have more than 4 clusters
for cluster_num in range(4):  # Change this if you chose a different number of clusters
    cluster_data = principal_df[principal_df['Cluster'] == cluster_num]
    plt.scatter(cluster_data['PC1'], cluster_data['PC2'], c=colors[cluster_num], label=f"Cluster {cluster_num}")

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.title("2D PCA of Customer Segments")
plt.title("2D PCA of Customer Segments")
plt.savefig('/Users/pintoza/Desktop/dev/data-science/sales_optimization/reports/models/customer_segmentation.png')
plt.show()
```
