```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/pintoza/Desktop/dev/data-science/sales_optimization/data/processed/df_processed.csv')

df.head()
```

```python
# Calculate rolling statistics
rolling_mean = df['unit_price'].rolling(window=12).mean()
rolling_std = df['unit_price'].rolling(window=12).std()

# Plot rolling statistics
sns.lineplot(x=df['order_date'], y=rolling_mean, label='Rolling Mean')
sns.lineplot(x=df['order_date'], y=rolling_std, label='Rolling Std Dev')
plt.legend()
plt.savefig('/Users/pintoza/Desktop/dev/data-science/sales_optimization/reports/eda_02/Rolling_Statistics.png')
```

```python
df['month'] = pd.DatetimeIndex(df['order_date']).month
sns.barplot(x='month', y='unit_price', data=df, estimator=sum)
plt.savefig('/Users/pintoza/Desktop/dev/data-science/sales_optimization/reports/Monthly_Average_Sales.png')
```

```python
df['year'] = pd.DatetimeIndex(df['order_date']).year
sns.lineplot(x='year', y='unit_price', data=df, estimator=sum)
plt.savefig('/Users/pintoza/Desktop/dev/data-science/sales_optimization/reports/Year_over_Year_Growth.png')
```
