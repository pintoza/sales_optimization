```python
# Create features
df['Month'] = pd.DatetimeIndex(df['order_date']).month
df['Quarter'] = pd.DatetimeIndex(df['order_date']).quarter
df['Discounted_Price'] = df['unit_price'] * (1 - df['discount_applied'])
df['Time_to_Delivery'] = (pd.DatetimeIndex(df['delivery_date']) - pd.DatetimeIndex(df['order_date'])).days
```

```python
# For Customer Segmentation
df['Customer_Lifetime_Value'] = df.groupby('customer_id')['unit_price'].transform('sum')

# For Product Recommendation
df['Product_Popularity'] = df.groupby('product_id')['order_number'].transform('count')

# For Sales Forecasting
df['Weekday'] = pd.DatetimeIndex(df['order_date']).weekday < 5

# For Inventory Optimization
df['Stock_Turnover_Ratio'] = df['order_quantity'] / df['order_quantity'].sum()
```
